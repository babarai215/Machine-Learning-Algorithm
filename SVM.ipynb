```python
# DROP TWO COLUMNS  'MONTH' AND 'DAY'
df4=df3.drop(columns=['month','day'],axis=1)
df4.head()
```

```python
df4.plot(kind='hist')
```

```python
# SPLITING DATA INTO DEPENDEDNT(TARGET) AND INDEPENDENT VARIABEL
y2 = df4['size_category']
x2 = df4.drop(['size_category'],axis=1)
```

# Random search cv

```python
# use random search cv for hyper parametet tuning 
parameters1 = {'kernel':['rbf'],
               'gamma':[100, 50, 5, 0.5, 0.1, 0.01, 0.0001, 'auto'],
               'C':[50, 15, 10, 6, 5, 0.1, 0.001, 0.0001]}

rsv1 = RandomizedSearchCV(model1, parameters1, cv=5,verbose = 3 ) # RandomizedSearchCV and gridSearchCV are same do work same but 
rsv1.fit(x_train2 , y_train2)                                       # romdam take randomly parameter but GridSearchCV take all 
                                                               # combination of parameter
    # if we want see what is going on iside model we have used verbose = 3
```

```python
# BUILD MODEL
model2 = SVC(C = 5, gamma = 0.0001)   # if C is less then margin high , if we take gama less underfit, take more then model  
model2.fit(x_train2 , y_train2)   
```

```python

y_predict1= model2.predict(x_test2)
y_predict1
```
